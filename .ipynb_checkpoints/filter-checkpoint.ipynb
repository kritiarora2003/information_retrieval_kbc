```python
import pandas as pd
import ast

def revert_dataframe_columns(df):
    """
    Attempts to convert stringified lists/dicts in all columns
    back to their original Python objects.
    """
    for col in df.columns:
        def try_eval(x):
            if isinstance(x, str):
                try:
                    return ast.literal_eval(x)
                except (ValueError, SyntaxError):
                    return x  # keep as string if it can't be evaluated
            return x
        df[col] = df[col].apply(try_eval)
    return df

revert_dataframe_columns(queries)
```

```python
import re

def clean_supressed_section_list(section_list):
    """
    Cleans each string in supressed_section list individually:
    - Removes citation numbers like [1], [2], etc.
    - Removes uppercase 'REFERENCE' and 'TARGETCASE' (with or without <>)
    - Removes 'supra'
    - Removes empty parentheses and extra punctuation
    - Strips extra spaces
    - Keeps only unique sentences per section
    """
    if not section_list:
        return []
    
    if not isinstance(section_list, list):
        section_list = [section_list]
    
    cleaned_list = []
    for text in section_list:
        # Remove citation numbers like [1], [23], etc.
        text = re.sub(r"\[\d+\]", "", text)
        # Remove uppercase REFERENCE/TARGETCASE (with or without < >)
        text = re.sub(r"<*REFERENCE>*|<*TARGETCASE>*", "", text)
        # Remove 'supra'
        text = re.sub(r"\bsupra\b", "", text, flags=re.IGNORECASE)
        # Remove empty parentheses
        text = re.sub(r"\(\s*\)", "", text)
        # Remove multiple punctuation marks and extra spaces
        text = re.sub(r"[;,\s]{2,}", " ", text)
        text = re.sub(r"\s+", " ", text)
        # Strip leading/trailing spaces
        text = text.strip()
        if text:  # only keep non-empty strings
            cleaned_list.append(text)
    
    # Remove duplicates within this section while preserving order
    seen = set()
    unique_list = []
    for item in cleaned_list:
        if item not in seen:
            seen.add(item)
            unique_list.append(item)
    
    return unique_list

    
queries["propositions"] = queries["suppressed_sections"].apply(clean_supressed_section_list)
queries
```

# bm25 running

```python
# Install BM25 library if not already
# pip install rank_bm25

import pandas as pd
from rank_bm25 import BM25Okapi
```

```python
# Suppose candidates_df is your candidate dataframe
candidate_texts = candidate_df['formatted_text'].tolist()

# Tokenize each candidate text
tokenized_candidates = [text.split() for text in candidate_texts]

# Initialize BM25 with all candidates
bm25 = BM25Okapi(tokenized_candidates)
```

```python
import time
import os
import pandas as pd

top_k = 50  # top 50 candidates per paragraph
results = []

output_file = "bm25_results_check.csv"

for idx, row in queries.head(800).iterrows():
    start_time = time.time()
    
    filename = row['filename']
    paras = row['propositions']  # list of paragraphs
    
    for para in paras:
        query_tokens = para.split()
        scores = bm25.get_scores(query_tokens)
        
        top_indices = scores.argsort()[-top_k:][::-1]
        
        # collect candidates and their scores for this paragraph
        top_candidates = candidate_df.iloc[top_indices]['candidate'].tolist()
        top_scores = [scores[i] for i in top_indices]
        
        results.append({
            'query_filename': filename,
            'paragraph': para,
            'candidates': top_candidates,   # list of top 50 candidates
            'scores': top_scores            # list of corresponding scores
        })
    
    end_time = time.time()
    elapsed = end_time - start_time
    
    # print time every alternate query
    if (idx + 1) % 2 == 0:
        print(f"Query {idx + 1} processed in {elapsed:.3f} seconds")
    
    # save after every 10 queries
    if (idx + 1) % 10 == 0:
        temp_df = pd.DataFrame(results)
        if os.path.exists(output_file):
            temp_df.to_csv(output_file, mode='a', index=False, header=False)
        else:
            temp_df.to_csv(output_file, index=False)
        print(f"Saved intermediate results after {idx + 1} queries")
        results = []  # clear to save memory

# save remaining results
if results:
    temp_df = pd.DataFrame(results)
    if os.path.exists(output_file):
        temp_df.to_csv(output_file, mode='a', index=False, header=False)
    else:
        temp_df.to_csv(output_file, index=False)
    print("Saved final remaining results!")

print("All queries processed!")
```
